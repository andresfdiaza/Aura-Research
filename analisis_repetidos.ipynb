# Análisis de datos repetidos en el CSV

Este notebook revisa si hay filas repetidas en el archivo `cv_datos_generales.csv` usando los campos `id_investigador`, `titulo_proyecto` y `nodo_padre`.

## 1. Importar librerías necesarias
Importamos pandas y otras librerías útiles para el análisis.

```python
import pandas as pd
import numpy as np
```

## 2. Cargar y explorar los datos
Cargamos el archivo `cv_datos_generales.csv` y mostramos las primeras filas para revisar la estructura.

```python
# Cargar el archivo CSV
ruta_csv = '../backend/cv_datos_generales.csv'
df = pd.read_csv(ruta_csv)
df.head()
```

# Seleccionar solo las columnas de interés

```python
# Seleccionar solo las columnas de interés
filtros = ['id_investigador', 'titulo_proyecto', 'nodo_padre']
df_filtros = df[filtros]
df_filtros.head()
```

## 4. Buscar y mostrar datos repetidos según los 4 filtros
Identificamos las filas duplicadas considerando solo los cuatro campos clave.

```python
# Buscar duplicados considerando los cuatro filtros
repetidos = df_filtros[df_filtros.duplicated(keep=False)]
repetidos.sort_values(by=filtros).head(10)
```

## 5. Contar y mostrar los datos únicos y repetidos
Calculamos cuántos registros son únicos y cuántos están repetidos según los cuatro filtros, mostrando ejemplos de ambos.

```python
# Cantidad de filas totales
total = len(df_filtros)
# Cantidad de filas únicas según los cuatro filtros
unicos = df_filtros.drop_duplicates().shape[0]
# Cantidad de filas repetidas
repetidos_cantidad = total - unicos

print(f"Total de filas: {total}")
print(f"Filas únicas (por los 4 filtros): {unicos}")
print(f"Filas repetidas (por los 4 filtros): {repetidos_cantidad}")

# Mostrar ejemplos de filas únicas y repetidas
print("\nEjemplo de filas repetidas:")
display(repetidos.head())

print("\nEjemplo de filas únicas:")
display(df_filtros.drop_duplicates().head())
```
